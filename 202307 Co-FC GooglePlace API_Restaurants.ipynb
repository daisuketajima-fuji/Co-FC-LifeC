```python
import googlemaps
import pprint #list型やDict型を見やすくprintするライブラリ
```

```python
# ここでGoogle Maps APIキーを設定します
key = 'AIzaSyB_VPV6jo29eM3pJJ_gCJrStr9ciSFKeGI'
# Google Maps APIを初期化します
client = googlemaps.Client(key)
```

```python
# 位置情報を検索
geocode_result = client.geocode('東京都江戸川区葛西5丁目')
# 軽度・緯度の情報のみ取り出す
loc = geocode_result[0]['geometry']['location']
```

```python
#半径200m以内のレストランの情報を取得
restaurant_result = client.places_nearby(location=loc, radius=200, type = 'food')
```

```python
pprint.pprint(restaurant_result)
```

```python
# ここから地図を表示する方法は以下のようになります (googlemapsの使用例)
import folium
```

```python
# 地図を初期化
map_center = [loc['lat'], loc['lng']]
map_restaurants = folium.Map(location=map_center, zoom_start=16)
```

```python
# レストランの位置を地図に追加
for restaurant in restaurant_result['results']:
    lat = restaurant['geometry']['location']['lat']
    lng = restaurant['geometry']['location']['lng']
    name = restaurant['name']
    folium.Marker([lat, lng], popup=name).add_to(map_restaurants)
```

```python
# 地図を表示
map_restaurants.save('restaurants_map.html')  # HTMLファイルに保存
```
